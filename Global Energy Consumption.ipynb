```python
# imports and setup

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv(r'C:\Users\1\Desktop\global_energy_consumption.csv')
```

```python
# data exploration

df.head()
df.describe()
df.isnull().sum()
```

```python
# gets the numeric datatype columns

ds_num = df.select_dtypes(include=['number'])
ds_num.corr()
```

```python
# visulization :
# distribution of total enery consumption

plt.figure(figsize = (10,6))
sns.histplot(df['Total Energy Consumption (TWh)'],kde=True)
```

```python
# visulization

# top10 contries by energy consumption

plt.figure(figsize = (15,20))
top_countries = df.groupby("Country")["Total Energy Consumption (TWh)"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
top_countries.plot(kind='bar', color='orange')
plt.title('Top 10 Countries by Energy Consumption')
plt.ylabel('Total Energy Consumption (TWh)')
plt.xticks(rotation=45)

# add quantity labels on each columns:

for i,consumption in enumerate(top_countries):
    plt.text(i,consumption,f'{consumption:,.0f}',ha = 'center',va = 'bottom')
plt.tight_layout()
plt.show()
```

```python
# visualization:
# correlations bewteen each features:

# auto select numeric columns
plt.figure(figsize =(10,6))
df_numeric = df.select_dtypes(include = ['number'])

corr_matrix = df_numeric.corr()
sns.heatmap(corr_matrix,annot = True,cmap = 'RdBu',fmt = '.2f')
```

```python
# visulization:
# Enery Consumption Trends Over The Year

yearly_total_consumption = df.groupby('Year')['Total Energy Consumption (TWh)'].sum()
yearly_total_consumption
plt.figure(figsize = (12,6))
yearly_total_consumption.plot(kind = 'line')
plt.title('Enery Consumption Trends Over The Year',fontsize = 16)
plt.ylabel('Total Energy Consumption (TWh)',fontsize = 13)
plt.xlabel('Year',fontsize = 13)
plt.show()
```

```python
# Visualization:
# Top10 Countries By Per Capital Energy Consumption

per_capital_energy_use = df.groupby('Country')['Per Capita Energy Use (kWh)'].sum().head(10)
sorted_per_capital_energy_use = per_capital_energy_use.sort_values(ascending=True)
plt.figure(figsize = (12,6))
sorted_per_capital_energy_use.plot(kind='bar', color='skyblue')
for i , y in enumerate (sorted_per_capital_energy_use):
        plt.text(i,y,f'{y:,.0f}',ha = 'center',va = 'bottom')

plt.title('Top 10 Countries by Per Capital Energy Consumption',fontsize = 16)
plt.xlabel('country',fontsize = 14)
plt.ylabel('Per Capital Consumption',fontsize = 14)
```

```python
# Visualization:

# Gets the number of records per country

plt.figure(figsize=(12, 6))
country_counts = df['Country'].value_counts()
sns.countplot(y='Country', data=df, order=country_counts.index)
plt.title('Number of Records Per Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
